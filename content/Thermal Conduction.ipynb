```python
from pint import UnitRegistry
p = UnitRegistry(autoconvert_offset_to_baseunit = True)
p.default_format = '.1f~P'
```





Source: Fundamentals of Heat and Mass Transfer, Incropera, et. al, 6th Edition

```python
k = 10 * (p.W/(p.m*p.K))
L = 1 * p.m
T1 = 200 * p.degC
T2 = 300 * p.degC
print(k)
print(L)
print(T1)
print(T2)
```

```python
q = -k * (T1-T2) / L
print(q.to(p.W/p.m**2))
```
