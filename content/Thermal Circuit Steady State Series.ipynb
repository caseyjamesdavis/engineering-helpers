```python
T1=150*p.degC
#T2=?
#T3=?
T4=10*p.degC

R1=(0.03*p.m)/(0.07*p.W/p.m/p.K)
R2=(0.1*p.m)/(0.7*p.W/p.m/p.K)
R3=R1

# print(R1)
# print(R2)
# print(R3)

q=(T1-T4)/(R1+R2+R3)
print(q.to(p.W/p.m**2))

T2=T1-q*R1
print(T2)

T3=T2-q*R2
print(T3)
```
